```python
country_name=final_df.Country.value_counts().index
```

```python
Country_count=final_df.Country.value_counts().values
```

```python
Country_count
```

```python
#top 3 countries 
plt.pie(Country_count[:3],labels=country_name[:3],autopct="%1.1f%%",
        shadow=True,startangle = 30,
        explode=(0.04,0.04,0.04))
plt.show()
```

```python
final_df.columns
```

```python
#Need to split some data for drive some important insights 
final_df.groupby(by=['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
#resetindex to convert to dataframe 
final_df.groupby(by=['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

# Observation: 
1. When Rating is between 4.5 to 4.9---> Excellent
2. When Rating are between 4.0 to 3.4--->very good
3. when Rating is between 3.5 to 3.9----> good
4. when Rating is between 3.0 to 3.4----> average
5. when Rating is between 2.5 to 2.9----> average
6. when Rating is between 2.0 to 2.4----> Poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
plt.show()
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings,hue="Rating color",palette=['blue','red','orange','yellow','green','green'])
```
